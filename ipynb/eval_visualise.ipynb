```python
def parallel_display(idx, figsize = (30, 15)):
    src = f'../test/hk_00{idx}.jfif'
    dst = f'../chkpts/first_blood/test/style_hk_00{idx}.jfif'
    src_img = Image.open(src)
    dst_img = Image.open(dst)
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize = figsize)
    ax1.imshow(src_img)
    ax2.imshow(dst_img)
    ax1.axis('off')
    ax2.axis('off')
    plt.tight_layout();
```

```python
parallel_display(1, figsize = (14, 7))
```
